```python
data_path = 'test'
file_name = 'MWC_Survey_clean.xlsx'

df = fn.read_data(data_path,file_name)

std = StandardScaler()
data = df[df.columns[:-3]]
X = std.fit_transform(data)
variable_names = list(data.columns)
xname = variable_names
xname.insert(0,'const')

X = pd.DataFrame(X, columns = data.columns)
X = sm.add_constant(X)
y = (df["tech_scenario"]/df["business_scenario"]).values

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
#print('X_train shape: ',np.shape(X_train))
#print('X_test shape: ',np.shape(X_test))
```

```python
max_cores = 90

test_size = 0.105
beta_vec = np.arange(2e-2,3.7e-2,0.2e-2)
k_folds = 10
n_runs = int(max_cores/k_folds)*k_folds

res = fn.different_models(X, y, k_folds, n_runs=n_runs, layers=[[20,10,5]], beta_vec=beta_vec, variable_names=variable_names)
```

```python
## OLS summary
print('base_loss_linear: {}'.format(res[0][3]))
print(res[0][0].summary(xname=xname))

## NN histograms
for result in res[1:]:
    print('base_loss: {}'.format(result[3]))
    print('r2: {}'.format(result[2]))
    fn.plot_heat_map(result[0], result[1], beta_vec=beta_vec)
```
